```python
December['trip_start_timestamp'] = pd.DatetimeIndex(December['trip_start_timestamp']).hour
December = December[December['trip_miles']>0.0]
December.reset_index(inplace=True)
December.drop('index',axis=1,inplace=True)
columns = ['trip_start_timestamp','trip_seconds','trip_miles',
           'pickup_community_area','dropoff_community_area','trip_total']
December_model = December.loc[:,columns]
for i in range(1001):
    if December_model.loc[i,'pickup_community_area'] == 11 or December_model.loc[i,'pickup_community_area'] == 1 or December_model.loc[i,'pickup_community_area'] == 2 or December_model.loc[i,'pickup_community_area'] == 3 or December_model.loc[i,'pickup_community_area'] == 4 or December_model.loc[i,'pickup_community_area'] == 5 or December_model.loc[i,'pickup_community_area'] == 6 or December_model.loc[i,'pickup_community_area'] == 7 or December_model.loc[i,'pickup_community_area'] == 10 or December_model.loc[i,'pickup_community_area'] == 22 or December_model.loc[i,'pickup_community_area'] == 8 or December_model.loc[i,'pickup_community_area'] == 16:
        December_model.loc[i,'Downtown'] = 1
    else :
        December_model.loc[i,'Downtown'] = 0
    i=i+1
for i in range(1001):
    if December_model.loc[i,'pickup_community_area'] == 11 or December_model.loc[i,'pickup_community_area'] == 1 or December_model.loc[i,'pickup_community_area'] == 2 or December_model.loc[i,'pickup_community_area'] == 3 or December_model.loc[i,'pickup_community_area'] == 14 or December_model.loc[i,'pickup_community_area'] == 5 or December_model.loc[i,'pickup_community_area'] == 6 or December_model.loc[i,'pickup_community_area'] == 7 or December_model.loc[i,'pickup_community_area'] == 13 or December_model.loc[i,'pickup_community_area'] == 24 or December_model.loc[i,'pickup_community_area'] == 8 :
        December_model.loc[i,'Famous'] = 1
    else :
        December_model.loc[i,'Famous'] = 0
    i=i+1
for i in range(1001):
    x = December_model.loc[i,'pickup_community_area']
    dat = December[December['pickup_community_area']==x].nunique()['company']
    December_model.loc[i,'no of companies'] = dat
    i = i+1
Dec = December_model[:1000]
```
